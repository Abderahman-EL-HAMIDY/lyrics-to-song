```python
import yt_dlp

def download_audio(video_url):
    # Options for downloading audio
    ydl_opts = {
        'format': 'bestaudio/best',  # Download best available audio quality
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'wav', # Convert to mp3
            'preferredquality': '192', # Audio bitrate (192, 256, 320, etc.)
        }],
        'outtmpl': '%(title)s.%(ext)s', # Name the file using the video title
        'quiet': False, # Print progress to console
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            print(f"Downloading: {video_url}")
            ydl.download([video_url])
            print("Download complete!")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example Usage
if __name__ == "__main__":
    url = "https://youtu.be/2KAFUzsR03M?si=FzFTF4lcSPQKyfmL"
    download_audio(url)
```

```python
from transformers import AutoProcessor, MusicgenMelodyForConditionalGeneration
import librosa
from IPython.display import Audio
import torch

# Load the MELODY-specific model (important!)
processor = AutoProcessor.from_pretrained("facebook/musicgen-melody")
model = MusicgenMelodyForConditionalGeneration.from_pretrained("facebook/musicgen-melody")

# Load your singing voice recording
audio_array, sampling_rate = librosa.load(
    "/kaggle/working/David Kushner — Daylight [ بدون موسيقى +مترجمة ].wav", 
    sr=32000
)

# Extract from 5 seconds to 15 seconds (10 seconds total)
start_sample = 0   # 5 seconds * 32000 = 160000
end_sample = 15 * sampling_rate    # 15 seconds * 32000 = 480000
audio_slice = audio_array[start_sample:end_sample]

print(f"Original audio length: {len(audio_array)/sampling_rate:.2f} seconds")
print(f"Sliced audio length: {len(audio_slice)/sampling_rate:.2f} seconds")

# Process both text and audio together
inputs = processor(
    audio=audio_slice,  # Use the sliced audio
    sampling_rate=sampling_rate,
    text=["pop ballad with piano, strings, and soft drums accompaniment"],
    padding=True,
    return_tensors="pt",
)

# Generate instrumentation around your voice
audio_values = model.generate(
    **inputs,
    do_sample=True, 
    guidance_scale=3,
    max_new_tokens=256  # ~20 seconds of output
)

# Play the result
sampling_rate = model.config.audio_encoder.sampling_rate
Audio(audio_values[0].cpu().numpy(), rate=sampling_rate)
```

```python
import numpy as np
from scipy.io import wavfile
from IPython.display import Audio, display

# Debug: Check what audio_values contains
print(f"audio_values type: {type(audio_values)}")
print(f"audio_values shape: {audio_values.shape}")

# Extract the generated audio correctly
if len(audio_values.shape) == 3:
    # Shape is likely [batch, channels, samples]
    generated_audio = audio_values[0, 0].cpu().numpy()  # Take first batch, first channel
elif len(audio_values.shape) == 2:
    # Shape is [batch, samples]
    generated_audio = audio_values[0].cpu().numpy()
else:
    # Shape is [samples]
    generated_audio = audio_values.cpu().numpy()

print(f"\nOriginal vocal length: {len(audio_slice)/sampling_rate:.2f} seconds ({len(audio_slice)} samples)")
print(f"Generated audio length: {len(generated_audio)/sampling_rate:.2f} seconds ({len(generated_audio)} samples)")
print(f"Generated audio shape: {generated_audio.shape}")

# Only proceed if we have generated audio
if len(generated_audio) == 0:
    print("\n❌ ERROR: Generated audio is empty!")
    print("Check your model generation - it may have failed or returned empty output")
else:
    # Mix the audio
    min_length = min(len(audio_slice), len(generated_audio))
    
    audio_slice_trimmed = audio_slice[:min_length]
    generated_audio_trimmed = generated_audio[:min_length]
    
    # Mix: 85% original vocals + 15% generated music
    vocal_volume = 0.6
    music_volume = 0.4
    mixed_audio = (vocal_volume * audio_slice_trimmed) + (music_volume * generated_audio_trimmed)
    
    # Normalize
    mixed_audio = mixed_audio / np.max(np.abs(mixed_audio))
    
    # Save the mixed portion
    mix_duration = len(mixed_audio) / sampling_rate
    wavfile.write(f'mixed_output_{mix_duration:.1f}sec.wav', sampling_rate, (mixed_audio * 32767).astype(np.int16))
    
    # If generated audio is shorter than 15 seconds, append remaining vocals
    if len(generated_audio) < len(audio_slice):
        remaining_vocals = audio_slice[min_length:]
        full_mixed_audio = np.concatenate([mixed_audio, remaining_vocals])
        wavfile.write('mixed_output_full.wav', sampling_rate, (full_mixed_audio * 32767).astype(np.int16))
        print(f"\n✅ Created full version: {len(full_mixed_audio)/sampling_rate:.1f} seconds")
    
    # Play all versions
    print("\n=== Original Vocals (15 sec) ===")
    display(Audio(audio_slice, rate=sampling_rate))
    
    print(f"\n=== Generated Music ({len(generated_audio)/sampling_rate:.1f} sec) ===")
    display(Audio(generated_audio, rate=sampling_rate))
    
    print(f"\n=== Mixed ({mix_duration:.1f} sec) ===")
    display(Audio(mixed_audio, rate=sampling_rate))
```
